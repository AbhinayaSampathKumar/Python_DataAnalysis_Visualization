```python
# USING THE PIVOT FUNCTION

```

```python
Smokers=Smoker.dropna()
# dropping all the null/nan values before pivot
```

```python
print(Smokers)
```

```python
Smoker1= Smokers.pivot_table(index=['gender', 'highest_qualification'] ,
                              values=['amt_weekdays', 'amt_weekends'],
                              aggfunc= 'mean')

# Round values for aggregate function
Smoker1=Smoker1.round(0)
print(Smoker1)
```

```python
# In the pivoted table the index columns will not be regular columns .To make it part of the data frame use the reset_index.
```

```python
Smoker1.reset_index(inplace=True)
```

```python
print(Smoker1.columns)
```

```python
# CONCAT FUNCTION-Concat the two tables.Append rows of dataframe
```

```python
Smoker_male = Smoker1[Smoker1['gender'] == 'Male'][['gender', 'highest_qualification', 'amt_weekdays', 'amt_weekends']]
Smoker_female = Smoker1[Smoker1['gender'] == 'Female'][['gender', 'highest_qualification', 'amt_weekdays', 'amt_weekends']]

# Concatenate the two DataFrames
concatenated = pd.concat([Smoker_female, Smoker_male])

print(concatenated)
```

```python
# Append columns of data frames

Smoker_male = Smoker1[Smoker1['gender'] == 'Male'][['gender','highest_qualification', 'amt_weekdays', 'amt_weekends']]
Smoker_female = Smoker1[Smoker1['gender'] == 'Female'][['gender', 'highest_qualification', 'amt_weekdays', 'amt_weekends']]


concatenated = pd.concat([Smoker_male[['gender',  'amt_weekdays', 'amt_weekends']], 
                          Smoker_female[['gender', 'amt_weekdays', 'amt_weekends']]], axis=1)

print(concatenated)


```
